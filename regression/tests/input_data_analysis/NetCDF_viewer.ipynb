```python
input_vars = ['qsm', 'qtm', 'qlm', 'skew_l', 'var_l', 'var_t', 'tm', 'pm','cl_l']
fname1    = 'ncr_pdf_douze_1deg.nc'
fname05   = 'ncr_pdf_douze_05deg.nc'
fname025  = 'ncr_pdf_douze_025deg.nc'
fname0125 = 'ncr_pdf_douze_0125deg.nc'
# ds contains 63 fields
fname=fname0125
ds1    = xr.open_dataset(data_folder+fname1)[input_vars]
ds05   = xr.open_dataset(data_folder+fname05)[input_vars]
ds025  = xr.open_dataset(data_folder+fname025)[input_vars]
ds0125 = xr.open_dataset(data_folder+fname0125)[input_vars]
```

```python
def plot_mean_profile(field):
    fvals=pmvals=ds[field].values
    mean_field_profile=[fvals[:,0,i].mean() for i in range(150)]
    fig=plt.figure(figsize=(12,9))
    plt.plot(mean_field_profile)
    plt.grid()
    plt.title('mean {} profile, averaged over all 61440 entries'.format(field),fontsize=18)
    plt.ylabel(field,fontsize=15)
    plt.xlabel('Vertical level',fontsize=15)
    plt.gca().invert_yaxis()
    
```
