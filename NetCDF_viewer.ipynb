```python
# lets look inside NetCDF input file
# the 00625deg means that the subdomains used for that file were 0.0625 of a degree large, so roughly 7x7 km
fname='ncr_pdf_douze_00625deg.nc'
filepath=data_folder+fname
# ds contains 63 fields
ds = xr.open_dataset(filepath)
ds
```

```python
ds.variables
```

```python
# lets select pm field - mean pressure
pmvals=ds.pm.values
pmvals.shape
```

```python
def plot_mean_profile(field):
    fvals=pmvals=ds[field].values
    mean_field_profile=[fvals[:,0,i].mean() for i in range(150)]
    fig=plt.figure(figsize=(12,9))
    plt.plot(mean_field_profile)
    plt.grid()
    plt.title('mean {} profile, averaged over all 61440 entries'.format(field),fontsize=18)
    plt.ylabel(field,fontsize=15)
    plt.xlabel('Vertical level',fontsize=15)
    plt.gca().invert_yaxis()
    
```

```python
plot_mean_profile('pm')
```

```python
plot_mean_profile('cl_l')
```

```python
plot_mean_profile('tm')
```
