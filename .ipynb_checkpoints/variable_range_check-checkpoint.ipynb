```python
 # DEFAULT PARAMETERS - global variables
curdir='/home/igor/UNI/Master Project/Script/Data/'
fname='ncr_pdf_douze_0125deg.nc'
goal_var = 'cl_l'
input_vars = ['qsm', 'qtm', 'qlm', 'skew_l', 'var_l', 'var_t', 'tm', 'pm']
add_vars = ['qvlm','qsm']
eval_fraction=0.6
regtype = 'decision_tree'
max_depth_in=None
```

```python
# DATA PREPROCESSING
prepro=cft.DataPrepro(curdir,fname,goal_var,input_vars,add_vars,eval_fraction)
# methods should be in this particular order!!
processed_data = prepro.get_processed_data()
```

```python
# REGRESSION
goalvar_pred, goalvar_eval = cft.regression(regtype,processed_data,max_depth_in)
```

```python
fig,ax=plt.subplots()
im=ax.hist2d(goalvar_eval, goalvar_pred, bins=100, norm=LogNorm(),cmap=plt.cm.jet)
ax.set_box_aspect(1)
fig.colorbar(im[3])
# im[0] is 2d array - bi-dimensional histogram of samples x and y
check_value=1000
total_cells=im[0].flatten().shape[0]
high_value_cells=im[0][im[0]>check_value].flatten().shape[0]
print('total n of pixels in 2d histogram={}, n of cells with values>{} is= {}'
      .format(total_cells,check_value,high_value_cells))
print('maximum value of pixel in 2d histogram ={}'.format(im[0].max()))
```
